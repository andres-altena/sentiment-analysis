```python
import pandas as pd
df_sent = pd.read_excel("../data/processed/sentiment.xlsx")
df_sent["Test_conclude"] -= 1
df_sent["DateTime"] = [datetime.strptime(x, "%Y-%m-%d %H:%M:%S.%f") for x in df_sent["DateTime"]]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib
import numpy as np

class Visualiser():
    #This class will be used to produce the visualisations
    def __init__(self, df):
        self.df = df
        
    def print_df(self):
        print(self.df)
        return
###################################################################################################
###################################################################################################        
    def filtering(self, person, start_date, end_date):
        #Do not want to alter the overall df!
        df_filter = self.df
        if (start_date != "earliest_date"):
            #Filter so that we discard dates before start_date
            start_date_filter = datetime.strptime(start_date, "%Y-%m-%d")
            df_filter = df_filter[df_filter["DateTime"] >= start_date_filter]
        if (end_date != "latest_date"):
            #Filter so that we discard dates after end_date
            end_date_filter = datetime.strptime(end_date, "%Y-%m-%d")
            df_filter = df_filter[df_filter["DateTime"] <= end_date_filter]
        if (person != "everyone"):
            #Filter for the messages that just have the person 
            df_filter = df_filter[df_filter["Title"] == "Person"]
        
        #Only return the important cols?
        df_filter = df_filter[["DateTime", "Test_conclude"]]
        
        return df_filter
###################################################################################################
###################################################################################################
    def set_coarseness(self, df_coarseness, coarseness):
        # Set the coarseness to either days, weeks, months, years
        if coarseness == "day":
            df_coarseness["Date"] = [x.date() for x in df_coarseness["DateTime"]]
        
        elif coarseness == "week":
            #Note that the +1 is such that all weeks are concentrated on monday - retain datetime
            #type and average over week
            df_coarseness["Date"] = [datetime.strptime(x.strftime("%G-%V")+"-1", "%G-%V-%u")
                                     for x in df_coarseness["DateTime"]]
                
        elif coarseness == "month":
            df_coarseness["Date"] = [datetime.strptime(x.strftime("%Y-%m"), "%Y-%m")
                                     for x in df_coarseness["DateTime"]]
            
        elif coarseness == "year":
            df_coarseness["Date"] = [datetime.strptime(x.strftime("%Y"), "%Y")
                                     for x in df_coarseness["DateTime"]]
        
        else:
            raise ValueError("Coarseness must be set to day/week/month/year")
        
        return df_coarseness
###################################################################################################
###################################################################################################       
    def variations_with_date(self, person = "everyone", start_date = "earliest_date",
                             end_date = "latest_date", coarseness = "day"):
        #Plots a graph of sentiment variations through dates
        #Date inputs must be in the form string: "YYYY-MM-DD"
        #
        #
        
        df_filter = self.filtering(person, start_date, end_date)
        
        #Groupby - would be nice to be able to vary the courseness of averaging
        df_coarseness = self.set_coarseness(df_filter, coarseness)
        
        
        df_filter_date = df_filter[["Date", "Test_conclude"]]
        #Groupby date, as_index false is important as it allows Date to be kept as date type
        gr = df_filter_date.groupby("Date", as_index = False)
        df_grouped = gr.agg({"Test_conclude": ["mean", "count"]})
        #df_grouped = df_grouped[df_grouped["Test_conclude"]["count"] > 5]
        
        #Plot the results:
        
        matplotlib.rc_file_defaults()
        ax1 = sns.set_style(style = None)
        fig, ax1 = plt.subplots(figsize=(16,8))
        
        gmean = sns.scatterplot(x = df_grouped["Date"], y = df_grouped["Test_conclude"]["mean"],
                     ax=ax1, marker = "o")#, hue = df_grouped.index.day)
        mean_yticks = np.arange(-1,1.25,0.25)
        gmean.set_yticks(mean_yticks)
        
        
        #Duplicate axis on same graph
        ax2 = ax1.twinx()     
        gcount = sns.lineplot(x = df_grouped["Date"], y = df_grouped["Test_conclude"]["count"],
                    alpha=0.5, ax=ax2, color = "red")
                
        count_yticks = np.linspace(0,round(max(df_grouped["Test_conclude"]["count"]),-2),5)
        plt.ylim(0,4*max(df_grouped["Test_conclude"]["count"]))
        plt.ylabel("         Number of texts", loc = "bottom")
        gcount.set_yticks(count_yticks)
                
        #plot = sns.lineplot(x = df_grouped.index, y = df_grouped["Test_conclude"]["mean"], markers = "x")
        #plot2 = sns.barplot(x = df_grouped.index, y = df_grouped["Test_conclude"]["count"])
        #plot.set(ylim=(-1, 1))
        
        return
###################################################################################################
###################################################################################################
    def variations_with_time(self, person = "everyone", start_date = "earliest_date",
                             end_date = "latest_date"):
        df_filter = filtering(person, start_date, end_date)
        #We want to average the sentiment by hour:
        #Plot the results:
        sns.lineplot(x = "X", y = "Y")      
        return
        
```

```python
visualiser = Visualiser(df_sent)
```
